## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
```

```
data.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# resposta da questão 1.1

col_faltantes = data.columns[data.isnull().any()]

for col in col_faltantes:
  print(data[col].isnull().sum())
```

```
linhas_nulas = data[data.isnull().any(axis=1)]
```

```
linhas_nulas.head(11)
```

```
print(linhas_nulas.shape[0]/data.shape[0]*100)
```

```
data = data.dropna()

data.shape
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
def analise_numerica_df (data: str, col: str):
  min = data[col].min()
  media = round(data[col].mean(),1)
  desv = round(data[col].std(),1)
  max = data[col].max()

  print(col)
  print(f"Min: {min}")
  print(f"Méd: {media}")
  print(f"Desvio: {desv}")
  print(f"Max: {max}\n")
```

```
# resposta da questão 1.2

for col in col_faltantes[:-1]:
  analise_numerica_df(data, col)
```

```
for col in col_faltantes[:-1]:
  data[f"{col}_std"] = data[col].apply(lambda v: (v - data[col].mean()) / ( data[col].std() ) )
```

```
data.head()
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3

especie = list(set(data['species']))
ilhas = list(set(data['island']))
sexo = list(set(data['sex']))

for esp in especie:
  data[f'{esp}_species_nom'] = data['species'].apply(lambda e: 1 if e == esp else 0)

data["sex_male_nom"] = data["sex"].apply(lambda m: 1 if m == "Male" else 0 )
data["sex_female_nom"] = data["sex"].apply(lambda f: 1 if f == "Female" else 0 )
```
